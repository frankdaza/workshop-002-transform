#### Read the data from the origin: MLA_100k.jsonlines file and see its structure

```python
import os
from dotenv import load_dotenv
import pandas as pd
import json

# Load environment variables
load_dotenv()

resources_path = os.getenv("RESOURCES_PATH")
print(f"RESOURCES_PATH: {resources_path}")

transformed_csv_filename = "transformed.csv"

if not resources_path:
    raise ValueError("RESOURCES_PATH no está definido en el .env")


mla_filename = "MLA_100k.jsonlines"
df_mla = pd.read_json(f"{resources_path}/{mla_filename}", lines=True)
```

```python
# Read only the first line
jsonlines_path = f"{resources_path}/{mla_filename}"
with open(jsonlines_path, 'r', encoding='utf-8') as f:
    first_line = next(f).strip()

# Parse the string as JSON
parsed_data = json.loads(first_line)

# Pretty-print the JSON to the console
print(json.dumps(parsed_data, indent=4, ensure_ascii=False))
```

#### Remove columns without any data, data with a lot of nulls values or data that I think is not relevant to know if a product is new or used

#### Make tranformations to Warranty field

```python
# Show warranty values and then transform them, fill NaN values with "No Warranty"
print(df_mla['warranty'].value_counts())
print("#" * 100)
print()

df_mla['warranty'] = df_mla['warranty'].fillna("No Warranty")

print()
print(df_mla['warranty'].value_counts())
print("#" * 100)
```

#### Remove another columns I think we can rid off

```python
df_mla.drop([
    'last_updated',
    'title',
    'base_price',
    'international_delivery_mode',
    'currency_id',
    'stop_time',
    'start_time',
    'initial_quantity',
    'sold_quantity',
    'available_quantity',
    'pictures'
], axis=1, inplace=True)
```

#### Create new transformed file structure

```python
# 1) Convert date_created to datetime just once
df_mla['date_created'] = pd.to_datetime(df_mla['date_created'], errors='coerce')

# 2) Extract components using datetime accessors (vectorized)
df_mla['date_year_created'] = df_mla['date_created'].dt.year
df_mla['date_month_created'] = df_mla['date_created'].dt.month
df_mla['date_day_created'] = df_mla['date_created'].dt.day
df_mla['date_hour_created'] = df_mla['date_created'].dt.hour
df_mla['date_minute_created'] = df_mla['date_created'].dt.minute
df_mla['date_second_created'] = df_mla['date_created'].dt.second

# 3) Extract seller_address dictionary fields once 
def extract_seller_address(address):
    if isinstance(address, dict):
        return pd.Series({
            'seller_address_longitude': address.get('longitude', None),
            'seller_address_latitude': address.get('latitude', None)
        })
    return pd.Series({'seller_address_longitude': None, 'seller_address_latitude': None})

df_seller = df_mla['seller_address'].apply(extract_seller_address)

# 4) Concatenate those new columns
df_mla = pd.concat([df_mla, df_seller], axis=1)

# 5) Build the final df_transformed:
df_transformed = df_mla[[
    'id',
    'condition',
    'warranty',
    'price',
    'accepts_mercadopago',
    'automatic_relist',
    'status',
    'seller_address_longitude',
    'seller_address_latitude',
    'date_year_created',
    'date_month_created',
    'date_day_created',
    'date_hour_created',
    'date_minute_created',
    'date_second_created'
]].copy()

# 6) Save as CSV
df_transformed.to_csv(
    f"{resources_path}/{transformed_csv_filename}",
    index=False,
    encoding="utf-8"
)

print(f"Transformed CSV file saved at: {transformed_csv_filename}")
```
